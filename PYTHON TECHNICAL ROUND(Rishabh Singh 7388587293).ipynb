# Task 1

```python
import pandas as pd
```

```python
import glob
file_list = glob.glob("*closePosition.csv")
```

### glob module is used to read all file ending with closePosition.csv

```python
df_list = []
for i in file_list[:-1]:
    df_list.append(pd.read_csv(i))
```

```python
res_df = pd.concat(
    df_list,
    axis=0,
    join="outer",
    ignore_index=True,
    keys=None,
    levels=None,
    names=None,
    verify_integrity=False,
    copy=True,
)
res_df
```

# Task 3

## Positive Streak 

```python
index_start = 0
index_end = 0
count = 0
streak_total = 0
flag = True
winninglistStreak = []
for i in range(len(res_df['Pnl'])):
    num = res_df['Pnl'][i]
    if num>0:
        if flag:
            index_start = i
        flag = False
        count += 1
        streak_total += num
    else:
        flag = True
        index_end = i
        winninglistStreak.append((index_start, index_end, streak_total, count))
        count = 0
        streak_total = 0
        
winninglistStreak
```

```python
from random import randint

def quicksort(array):
    if len(array) < 2:
        return array

    low, same, high = [], [], []

    pivot = array[randint(0, len(array) - 1)]

    for item in array:
        if item[2] > pivot[2]:
            low.append(item)
        elif item[2] == pivot[2]:
            same.append(item)
        elif item[2] < pivot[2]:
            high.append(item)

    return quicksort(low) + same + quicksort(high)
```

```python
Winning = quicksort(winninglistStreak)
Winning
```

```python
from datetime import datetime
for i in Winning:
    datetime_start = datetime.strptime(res_df['Date'][i[0]], '%Y-%m-%d')
    datetime_end = datetime.strptime(res_df['Date'][i[1]], '%Y-%m-%d')
    print("No of Trades", "Start-End", "Pnl of streak")
    print(i[-1],",", datetime_end.date()-datetime_start.date(), ",", i[2])
```

```python
Final_win = ["No of Trades, Start-End, Pnl of streak"]

for i in Winning:
    Final_win.append("{} Trades  {} to {} Profit = {}".format(i[-1], res_df['Date'][i[0]], res_df['Date'][i[1]],i[2]))
Final_win
```

## negative streak

```python
index_start = 0
index_end = 0
count = 0
streak_total = 0
flag = True
LoosinglistStreak = []
for i in range(len(res_df['Pnl'])):
    num = res_df['Pnl'][i]
    if num<=0:
        if flag:
            index_start = i
        flag = False
        count += 1
        streak_total += num
    else:
        flag = True
        index_end = i
        LoosinglistStreak.append((index_start, index_end, streak_total, count))
        count = 0
        streak_total = 0
        
LoosinglistStreak
```

```python
from random import randint

def quicksort(array):
    if len(array) < 2:
        return array

    low, same, high = [], [], []

    pivot = array[randint(0, len(array) - 1)]

    for item in array:
        if item[2] < pivot[2]:
            low.append(item)
        elif item[2] == pivot[2]:
            same.append(item)
        elif item[2] > pivot[2]:
            high.append(item)

    return quicksort(low) + same + quicksort(high)
```

```python
Loosing = quicksort(LoosinglistStreak)
Loosing
```

```python
Final_Loos = ["No of Trades, Start-End, Pnl of streak"]

for i in Loosing:
    Final_Loos.append("{} Trades  {} to {} Profit = {}".format(i[-1], res_df['Date'][i[0]], res_df['Date'][i[1]],i[2]))
Final_Loos
```
